## Line Plot
Time Series Data

```python
import matplotlib.pyplot as plt
import numpy as np

yrs = [1900, 1925, 1970, 2000]
pop = [2.5, 3.5, 5, 6.9]

plt.plot(yrs, pop) # yrs = x-axis and pop = y-axis
```

## Scatterplot
This will be ideal to go for most of the datasets

```python
plt.scatter(yrs, pop)
```

## Histogram
Data Exploration
Get idea about distribution


```python
x = np.random.randn(10)
plt.hist(x, bins = 10)
# Or can also specify number of bins and other arguments 
# check out Help function for that
```

## Customization

##### Adding Labels

```python
plt.plot(yrs, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
```

#### Shifting Axis
Here, starting Y-axis from 0 in multiple of 2 using .yticks() function

```python
plt.plot(yrs, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.yticks([0, 2, 4, 6, 8, 10])
# Now, the curve shifts since we changed Y-axis
```

#### Naming Axis Numbers Properly

```python
plt.plot(yrs, pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.yticks([0, 2, 4, 6, 8, 10], ['0B', '2B', '4B', '6B', '8B', '10B'])
# Now you see changes in the X-axis labels
```
